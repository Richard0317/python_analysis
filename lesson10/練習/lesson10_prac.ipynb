# DictReader

# 無型別問題，但還是字串

```python
import csv

target_column_name = 'pm2.5'  # 替換成你要的標題名稱
values = []

with open('aqx_p_488.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)  # 讀取標題列
    # 找出目標欄位的索引
    try:
        col_index = header.index(target_column_name)
    except ValueError:
        raise Exception(f"找不到欄位名稱: {target_column_name}")

    # 讀取該欄位底下所有資料（標題列已跳過）
    for row in reader:
        if len(row) > col_index: # 應是確定每列完整
            values.append(row[col_index])

print(values)
```

# 在某範圍，已是數值

```python
import csv

target_column = 'pm2.5'
min_value = 0
max_value = 3
filtered_values = []

with open('aqx_p_488.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        try:
            val = float(row[target_column])
            if min_value <= val <= max_value:
                filtered_values.append(val)
        except (ValueError, KeyError):
            pass

print(filtered_values)
```

# pm2.5是第幾欄

```python
import csv

with open('aqx_p_488.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    
    # 取得欄位名稱列表（依讀取順序）
    fieldnames = reader.fieldnames
    
    print("欄位名稱列表:", fieldnames)
    
    # 假設要找的欄位名稱
    target_col = 'pm2.5'
    
    if target_col in fieldnames:
        index = fieldnames.index(target_col)
        print(f"欄位 '{target_col}' 是第 {index} 欄（從0開始計算）")
    else:
        print(f"欄位 '{target_col}' 不存在")
```
